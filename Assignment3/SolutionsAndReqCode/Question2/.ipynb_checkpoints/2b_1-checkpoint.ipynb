```python
# IMPORTS
import json
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# MACRO DEFINITIONS
analysis_data_centers = ['52.95.127.226 AWS Dublin', '8.208.40.164 Alibaba London', '47.254.186.91 Alibaba Frankfurt', '47.88.111.29 Alibaba SanJose', '103.8.174.238 AWS Sydney', '23.248.185.157 UCloud Lagos', '23.248.184.34 UCloud Dubai', '176.32.118.30 AWS SanFrancisco']
analysis_data_tags = ['ETHERNET', 'LTE', 'SAT', 'WIFI']
```

```python
# CREATE ANALYSIS GRAPHS
for each_data_center in analysis_data_centers:
    data_center_name = each_data_center.split(' ')[1] + ' -- ' + each_data_center.split(' ')[2]
    
    for each_data_tag in range(len(analysis_data_tags)):
        timestamp_rtts = []
        try: 
            with open('./Question2_Data/' + each_data_center.split(' ')[0] + '_' + analysis_data_tags[each_data_tag] + '.json', 'r') as file:
                json_data = json.load(file)

                for each_probe_result in json_data:
                    timestamp_rtts.append((each_probe_result['timestamp'], each_probe_result['avg']))
        except Exception as e:
            continue
        
        merged_timestamp_rtt_dict = {}
        for each_timestamp_rtt_pair in timestamp_rtts:
            key = each_timestamp_rtt_pair[0]
            value = each_timestamp_rtt_pair[1]
            if key in merged_timestamp_rtt_dict:
                merged_timestamp_rtt_dict[key][0] += value
                merged_timestamp_rtt_dict[key][1] += 1
            else:
                merged_timestamp_rtt_dict[key] = [value, 1]
        result_list = [(key, total_sum / count) for key, (total_sum, count) in merged_timestamp_rtt_dict.items()]
        
        timestamps = []
        rtts = []
        for timestamp, average_rtt in merged_timestamp_rtt_dict.items():
            timestamps.append()
        
        print(result_list)
    
#     bar_width = 0.4
#     bar_positions = range(len(categories))
    
#     plt.bar(bar_positions, max_values, width=bar_width, color='lightblue', label='Max RTT')
#     plt.bar(bar_positions, min_values, width=bar_width, color='lightcoral', label='Min RTT')
    
#     plt.plot(categories, avg_values, marker='o', color='green', label='Average RTT')
    
#     plt.xlabel('Tags')
#     plt.ylabel('RTT (ms)')
#     plt.title('RTT Values for datacenter ' + data_center_name + '(' + each_data_center.split(' ')[0] + ")")
    
#     plt.legend()

#     plt.show()

```
